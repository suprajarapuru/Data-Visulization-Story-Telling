```python
## Import necessary packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Load data to data frame
world_pop_df = pd.read_excel('C:/Users/Supraja/Desktop/Supraja WORK Apps/8. DSC640/Data/world-population.xlsm')

## Display data frame
world_pop_df
```

### Create line chart of population by year

```python
## Figure size parameter
plt.rcParams["figure.figsize"] = (20,10)

## Line chart of year versus population
plt.plot(world_pop_df['Year'], world_pop_df['Population'])

## Titles and labels
plt.title('Population by Year')
plt.xlabel('Year')
plt.ylabel('Population (billions)')

plt.show()
```

### Create step chart of population by year

```python
plt.plot(world_pop_df['Year'], world_pop_df['Population'], drawstyle='steps')
## Titles and labels
plt.title('Population by Year')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```
