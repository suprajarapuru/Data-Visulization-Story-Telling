```r
## Set directory and load file to data frame
library('ggplot2')
library('readxl')
setwd("C:/Users/Supraja/Desktop/Supraja WORK Apps/8. DSC640/Data")
library('tidyverse')
theme_set(theme_minimal())
```

```r
## Load data to data frame
world_pop_df <- read_excel("world-population.xlsm")
world_pop_df
```

### Create line chart

```r
ggplot(data=world_pop_df, aes(x=Year, y=Population)) + geom_line() + ggtitle('Population by Year') + xlab('Year') + ylab('Population (billions)')
```

### Create Step chart

```r
ggplot(data=world_pop_df, aes(x=Year, y=Population)) + geom_step() + ggtitle('Population by Year') + xlab('Year') + ylab('Population (billions)')
```
